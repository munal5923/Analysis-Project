```python
def course_col(uni_data, uni_course):
    full_data = []

    for i, row in uni_data.iterrows():
        institution_name = row['Institution name']
        
        if institution_name in uni_course.columns:
            courses = uni_course[institution_name].dropna()
            if len(courses) == 0:
                new_row = row.copy()
                new_row['Course'] = pd.NA
                full_data.append(new_row)
                
            for course in courses:
                new_row = row.copy()
                new_row['Course'] = course
                full_data.append(new_row)
        else:
            new_row = row.copy()
            new_row['Course'] = pd.NA
            full_data.append(new_row)
    
    result_df = pd.DataFrame(full_data)
    return result_df
```

```python
result_df = course_col(uni_data, uni_course)
result_df
```

```python
# Initialize an empty list to store intermediate data frames
dfs = []

# Iterate over each column in uni_course.columns
for col in uni_course.columns:
    data = result_df[result_df['Institution name'] == col]
    df_cleaned = data.drop_duplicates(subset=['Course'])
    dfs.append(df_cleaned)

# Concatenate all the filtered data frames into a single data frame
result_df_cleaned = pd.concat(dfs, ignore_index=True)

# Verify the result
result_df_cleaned
```
